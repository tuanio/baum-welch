```python
import numpy as np
import matplotlib.pyplot as plt
from bkt import BKT
from pprint import pprint
```

```python
def watch(bkt, idx):
    pprint(obs[idx])
    pprint(bkt.data_[idx])
    bkt.plot_posterior(idx, obs[idx])
```

```python
T = 15
M = 20
M, T
obs = np.random.choice([0, 1], size=(M, T))
obs
```

Each row is one student-skill observation with 15 tried

```python
%%time

bkt_self = BKT(random_state=1)
bkt_self.fit(obs)
```

```python
%%time

bkt_hmm = BKT(random_state=1, algorithm='hmm')
bkt_hmm.fit(obs)
```

```python
watch(bkt_hmm, 0)
```
